```python
import pandas as pd
```

```python
df_red = pd.read_csv('Red_Weather.csv')
df_rose = pd.read_csv('Rose_Weather.csv')
df_sparkling = pd.read_csv('Sparkling_Weather.csv')
df_white = pd.read_csv('White_Weather.csv')
```

```python
df_red['type'] = 'Red'
df_rose['type'] = 'Rose'
df_sparkling['type'] = 'Sparkling'
df_white['type'] = 'White'
```

```python
df_red.drop(columns=['city', 'country'], inplace=True)
```

```python
wine_df = pd.concat([df_red, df_rose, df_sparkling, df_white], axis=0, ignore_index=True)
wine_df
```

```python
#Let's standardize the column names
wine_df.columns = wine_df.columns.str.lower()
wine_df.rename(columns={'numberofratings':'num_ratings'}, inplace=True)
wine_df
```

```python
duplicates = wine_df[wine_df.duplicated()]
print(duplicates)
```

```python
wine_df.to_csv('wine_weather.csv', index=False)
```
