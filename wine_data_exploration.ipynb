```python
import pandas as pd
```

```python
df_red = pd.read_csv('Red.csv')
df_rose = pd.read_csv('Rose.csv')
df_sparkling = pd.read_csv('Sparkling.csv')
df_white = pd.read_csv('White.csv')
```

```python
df_red['type'] = 'Red'
df_rose['type'] = 'Rose'
df_sparkling['type'] = 'Sparkling'
df_white['type'] = 'White'
```

```python
wine_df = pd.concat([df_red, df_rose, df_sparkling, df_white], axis=0, ignore_index=True)
wine_df
```

```python
#Let's standardize the column names
wine_df.columns = wine_df.columns.str.lower()
wine_df.rename(columns={'numberofratings':'num_ratings'}, inplace=True)
wine_df
```

```python
duplicates = wine_df[wine_df.duplicated()]
print(duplicates)
```
